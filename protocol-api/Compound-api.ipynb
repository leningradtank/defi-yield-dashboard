```python
import requests
import numpy as np
import pandas as pd

def get_compound():

    data_apy = {'name': [], 'symbol': [], 'apy_borrow': [], 'apy_lending': []}
    res_draft = requests.get(f'https://api.compound.finance/api/v2/ctoken').json()
    res = res_draft['cToken']
    
    for i in range(len(res)):

        data_apy['name'].append(res[i]['name'])
        data_apy['symbol'].append(res[i]['underlying_symbol'])
        data_apy['apy_borrow'].append(round(float(res[i]['borrow_rate']['value'])*100,2))
        data_apy['apy_lending'].append(round(float(res[i]['supply_rate']['value'])*100, 2))
    
    df = pd.DataFrame(data_apy)
    display(df)
    print(res)

get_compound()
```
