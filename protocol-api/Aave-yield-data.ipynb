```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import time as time
import numpy as np
```

```python
#https://thegraph.com/hosted-service/subgraph/uniswap/uniswap-v3
#Subgraph pulling data from
def run_query(query): 
    request = requests.post('https://api.thegraph.com/subgraphs/name/aave/protocol-v2',
                            json={'query': query})
    print(request)
    if request.status_code == 200:
        return request.json()
    else:
        raise Exception('Query failed and return code is {}.      {}'.format(request.status_code,
                        query))
```

```python
#Data for feeTier, spot price, and liquidity  of USDC WETH Liquidity Pool
query ="""
{
  reserves {
    name
    underlyingAsset
    
    liquidityRate 
    stableBorrowRate
    variableBorrowRate
    
    aEmissionPerSecond
    vEmissionPerSecond
    sEmissionPerSecond
    
    totalATokenSupply
    totalCurrentVariableDebt
  }
}
"""
result = run_query(query)  # Execute the query

pd.json_normalize(result['data']['reserves'])
```

```python
df = pd.DataFrame()
```

```python
d1 = pd.json_normalize(result['data']['reserves'])
```

```python
d1.head()
```

```python
aave_data = pd.DataFrame()
aave_data['name'] = d1['name']
aave_data['liquidityRate'] = d1['liquidityRate'].astype(float)
aave_data['stableBorrowRate'] = d1['stableBorrowRate'].astype(float)
aave_data['variableBorrowRate'] = d1['variableBorrowRate'].astype(float)
aave_data.head(10)
```

```python
#Declaring some variables for Aave calculation 
RAY = 10**27 
SECONDS_PER_YEAR = 31536000
```

```python
aave_data['depositAPR'] = aave_data['liquidityRate']/RAY
aave_data['variableBorrowAPR'] = aave_data['variableBorrowRate']/RAY
aave_data['stableBorrowAPR'] = aave_data['stableBorrowRate']/RAY
```

```python
aave_data['depositAPY'] = (1 + (aave_data['depositAPR']/SECONDS_PER_YEAR)*SECONDS_PER_YEAR) - 1
aave_data['variableBorrowAPY'] = (1 + (aave_data['variableBorrowAPR']/SECONDS_PER_YEAR)*SECONDS_PER_YEAR) - 1
aave_data['stableBorrowAPY'] = (1 + (aave_data['stableBorrowAPR']/SECONDS_PER_YEAR)*SECONDS_PER_YEAR) - 1
```
